```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Sample - Superstore.csv", encoding="latin-1")
df.head()
```

```python
# 1. Chuyển Order Date sang datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
# 2. Tạo cột Year và Month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.to_period('M')
# 3. Tổng Sales và Profit theo từng tháng
monthly = df.groupby('Month')[['Sales','Profit']].sum().reset_index()
# Đổi month sang string
monthly['Month'] = monthly['Month'].astype(str)
print(monthly.head(10))
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Đường Sales
fig.add_trace(go.Scatter(
    x=monthly["Month"],
    y=monthly["Sales"],
    name="Sales",
    line=dict(color="royalblue")
))

# Đường Profit
fig.add_trace(go.Scatter(
    x=monthly["Month"],
    y=monthly["Profit"],
    name="Profit",
    line=dict(color="green")
))

fig.update_layout(
    title="Doanh thu & Lợi nhuận theo tháng",
    xaxis_title="Tháng",
    yaxis_title="USD",
    xaxis=dict(tickangle=45),  # Xoay label cho dễ đọc
    hovermode="x unified"       # Hover thấy cả 2 giá trị cùng lúc
)

fig.show()
```

```python
# Profit theo Category (nhóm lớn)
category = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()

# Tính thêm Profit Margin (%) = Profit / Sales * 100
# Ý nghĩa: cứ 100$ doanh thu thì lời bao nhiêu $
category["Profit Margin %"] = (category["Profit"] / category["Sales"] * 100).round(2)

print(category)
print("---")

# Profit theo Sub-Category (nhóm nhỏ hơn)
sub_cat = df.groupby("Sub-Category")[["Sales", "Profit"]].sum().reset_index()
sub_cat["Profit Margin %"] = (sub_cat["Profit"] / sub_cat["Sales"] * 100).round(2)

# Sắp xếp từ lỗ nhất đến lời nhất
sub_cat = sub_cat.sort_values("Profit", ascending=True)

print(sub_cat)
```

```python
import plotly.express as px

# Bar chart Sub-Category theo Profit
fig = px.bar(
    sub_cat,
    x="Profit",
    y="Sub-Category",
    orientation="h",           # Nằm ngang cho dễ đọc
    color="Profit",            # Tự động đổi màu: đỏ = âm, xanh = dương
    color_continuous_scale="RdYlGn",  # Red → Yellow → Green
    title="Profit theo Sub-Category"
)

fig.update_layout(yaxis={"categoryorder": "total ascending"})
fig.show()
```

```python
# Profit và Sales theo Region
region = df.groupby("Region")[["Sales", "Profit"]].sum().reset_index()
region["Profit Margin %"] = (region["Profit"] / region["Sales"] * 100).round(2)
region = region.sort_values("Profit", ascending=False)

print(region)
```

```python
fig = px.bar(
    region,
    x="Region",
    y="Profit",
    color="Profit Margin %",
    color_continuous_scale="RdYlGn",
    title="Profit theo Region",
    text="Profit Margin %"        # Hiện % ngay trên cột
)

fig.update_traces(texttemplate="%{text}%", textposition="outside")
fig.show()
```

```python
# Tạo nhóm discount để so sánh
# Ví dụ: 0%, 1-20%, 21-40%, 40%+
df["Discount Group"] = pd.cut(
    df["Discount"],
    bins=[-0.01, 0, 0.2, 0.4, 1.0],
    labels=["No Discount", "Low (1-20%)", "Medium (21-40%)", "High (40%+)"]
)

discount = df.groupby("Discount Group", observed=True)[["Sales", "Profit"]].mean().reset_index()
discount["Profit Margin %"] = (discount["Profit"] / discount["Sales"] * 100).round(2)

print(discount)
```

```python
fig = px.bar(
    discount,
    x="Discount Group",
    y="Profit Margin %",
    color="Profit Margin %",
    color_continuous_scale="RdYlGn",
    title="Profit Margin % theo mức Discount",
    text="Profit Margin %"
)

fig.update_traces(texttemplate="%{text}%", textposition="outside")
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Đường Sales
fig.add_trace(go.Scatter(
    x=monthly["Month"],
    y=monthly["Sales"],
    name="Sales",
    line=dict(color="#1f77b4", width=2)
))

# Đường Profit
fig.add_trace(go.Scatter(
    x=monthly["Month"],
    y=monthly["Profit"],
    name="Profit",
    line=dict(color="#2ca02c", width=2)
))

# Vẽ đường kẻ ngang tại y=0 để dễ thấy profit âm
fig.add_hline(
    y=0,
    line_dash="dash",
    line_color="red",
    annotation_text="Break-even",
    annotation_position="bottom right"
)

# Annotation cho 2 tháng bất thường
fig.add_annotation(
    x="2014-07", y=-841,
    text="⚠️ Profit âm",
    showarrow=True,
    arrowhead=2,
    font=dict(color="red")
)

fig.add_annotation(
    x="2015-01", y=-3281,  # Thay bằng giá trị thực tế của bạn
    text="⚠️ Profit âm",
    showarrow=True,
    arrowhead=2,
    font=dict(color="red")
)

fig.update_layout(
    title=dict(text="Doanh thu & Lợi nhuận theo tháng (2014–2017)", font=dict(size=16)),
    xaxis_title="Tháng",
    yaxis_title="USD ($)",
    hovermode="x unified",
    plot_bgcolor="white",
    legend=dict(orientation="h", y=1.1)  # Legend nằm ngang phía trên
)

fig.show()
```

```python
fig2 = px.bar(
    sub_cat,
    x="Profit",
    y="Sub-Category",
    orientation="h",
    color="Profit",
    color_continuous_scale="RdYlGn",
    title="Lợi nhuận theo Sub-Category",
    text="Profit"
)

fig2.update_traces(
    texttemplate="$%{text:,.0f}",   # Format số có dấu phẩy + ký hiệu $
    textposition="outside"
)

fig2.update_layout(
    plot_bgcolor="white",
    xaxis_title="Profit ($)",
    yaxis_title="",                  # Không cần label trục Y vì đã có tên
    coloraxis_showscale=False,       # Ẩn thanh màu bên phải cho gọn
    title=dict(font=dict(size=16)),
    xaxis=dict(tickformat="$,.0f")   # Format trục X theo $
)

fig2.show()
```

```python
fig3 = px.bar(
    region,
    x="Region",
    y="Profit",
    color="Profit Margin %",
    color_continuous_scale="RdYlGn",
    title="Lợi nhuận theo Region",
    text="Profit Margin %"
)

fig3.update_traces(
    texttemplate="%{text}%",
    textposition="outside"
)

fig3.update_layout(
    plot_bgcolor="white",
    xaxis_title="Region",
    yaxis_title="Profit ($)",
    coloraxis_showscale=False,
    title=dict(font=dict(size=16)),
    yaxis=dict(tickformat="$,.0f")
)

fig3.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig_combined = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Doanh thu & Lợi nhuận theo tháng",
        "Lợi nhuận theo Sub-Category",
        "Lợi nhuận theo Region",
        "Profit Margin % theo mức Discount"
    )
)

# --- Chart 1: Monthly Sales & Profit ---
fig_combined.add_trace(
    go.Scatter(x=monthly["Month"], y=monthly["Sales"],
               name="Sales", line=dict(color="#1f77b4")),
    row=1, col=1
)
fig_combined.add_trace(
    go.Scatter(x=monthly["Month"], y=monthly["Profit"],
               name="Profit", line=dict(color="#2ca02c")),
    row=1, col=1
)

# --- Chart 2: Sub-Category — số nằm trong bar ---
fig_combined.add_trace(
    go.Bar(
        x=sub_cat["Profit"],
        y=sub_cat["Sub-Category"],
        orientation="h",
        marker=dict(color=sub_cat["Profit"], colorscale="RdYlGn"),
        text=sub_cat["Profit"].apply(lambda x: f"${x:,.0f}"),
        textposition="inside",
        showlegend=False
    ),
    row=1, col=2
)

# --- Chart 3: Region — % nằm trong bar ---
fig_combined.add_trace(
    go.Bar(
        x=region["Region"],
        y=region["Profit"],
        marker=dict(color=region["Profit Margin %"], colorscale="RdYlGn"),
        text=region["Profit Margin %"].apply(lambda x: f"{x}%"),
        textposition="inside",
        showlegend=False
    ),
    row=2, col=1
)

# --- Chart 4: Discount — bỏ số, trục Y đủ rõ ---
fig_combined.add_trace(
    go.Bar(
        x=discount["Discount Group"].astype(str),
        y=discount["Profit Margin %"],
        marker=dict(color=discount["Profit Margin %"], colorscale="RdYlGn"),
        showlegend=False
    ),
    row=2, col=2
)

fig_combined.update_layout(
    title=dict(
        text="Sales Performance Dashboard — Superstore 2014–2017",
        font=dict(size=20)
    ),
    height=1200,
    plot_bgcolor="white",
    paper_bgcolor="white"
)

fig_combined.show()
```
